```python
# Function to detect anomalies using IQR
def detect_anomalies(df, column):
    # Calculate Q1, Q3, and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Flag anomalies
    df[f'{column}_Anomaly'] = (df[column] < lower_bound) | (df[column] > upper_bound)
    
    return df

# Detect anomalies in temperature and precipitation
df = detect_anomalies(df, 'T_Q')
df = detect_anomalies(df, 'PRELIQ_Q')

# Display rows where anomalies were detected in either temperature or precipitation
anomalies = df[(df['T_Q_Anomaly'] == True) | (df['PRELIQ_Q_Anomaly'] == True)]
print(anomalies[['DATE', 'T_Q', 'PRELIQ_Q', 'T_Q_Anomaly', 'PRELIQ_Q_Anomaly']])
```

```python
# Adjusted function to detect anomalies with a 3x IQR threshold
def detect_anomalies(df, column, multiplier=3):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    
    anomalies_mask = (df[column] < lower_bound) | (df[column] > upper_bound)
    df[f'{column}_Anomaly'] = anomalies_mask
    
    # Return only the anomalies
    return df[df[f'{column}_Anomaly'] == True]

# Detecting temperature anomalies with a stricter threshold
temp_anomalies = detect_anomalies(df, 'T_Q', multiplier=3)

# Detecting precipitation anomalies with a stricter threshold
precip_anomalies = detect_anomalies(df, 'PRELIQ_Q', multiplier=3)

# Print the shapes of the anomaly dataframes as a sanity check
print(f'Temperature Anomalies: {temp_anomalies.shape}')
print(f'Precipitation Anomalies: {precip_anomalies.shape}')

# You can then explore or export these tables for further analysis
```

```python
from sklearn.ensemble import IsolationForest

# Function to apply Isolation Forest on a grouped subset of the data
def apply_isolation_forest(group):
    # Assuming 'T_Q' is the column to check for anomalies
    if len(group) > 1:  # Ensuring there are enough data points for the model
        model = IsolationForest(n_estimators=100, contamination=0.00001, random_state=42)
        X = group[['T_Q']]  # Feature matrix
        group['Anomaly_IsolationForest'] = model.fit_predict(X)
    else:
        group['Anomaly_IsolationForest'] = 1  # Mark as normal if only one data point
    return group

# Group the data by month and day, then apply the Isolation Forest
grouped = df.groupby(['month', 'day'])
results = grouped.apply(apply_isolation_forest)

# Filter to see the anomalies
anomalies = results[results['Anomaly_IsolationForest'] == -1]

print(anomalies[['LAMBX', 'LAMBY', 'DATE', 'T_Q', 'Anomaly_IsolationForest']])
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Assuming 'T_Q' is the column for temperature
X = df[['T_Q']].to_numpy()  # Convert the temperature column to a numpy array

# Initialize and fit DBSCAN
dbscan = DBSCAN(eps=3, min_samples=2).fit(X)

# Labels of the clusters found by DBSCAN. -1 indicates outliers
df['Anomaly_DBSCAN'] = dbscan.labels_

# Filter out the anomalies for review
anomalies_dbscan = df[df['Anomaly_DBSCAN'] == -1]

print(anomalies_dbscan[['DATE', 'T_Q', 'Anomaly_DBSCAN']])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Sélection des caractéristiques
features = df[['T_Q', 'PRELIQ_Q', 'HU_Q']]  # Température, précipitations liquides, humidité relative

# Standardisation des caractéristiques
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Application de K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(features_scaled)
df['Cluster'] = clusters

# Analyse des Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='T_Q', y='PRELIQ_Q', hue='Cluster', palette='viridis')
plt.title('Clusters de Conditions Climatiques')
plt.xlabel('Température Moyenne (°C)')
plt.ylabel('Précipitations Liquides (mm)')
plt.legend(title='Cluster')
plt.show()
```
